```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("processing_results.csv", sep = ",")
results
```

```python
area_plot = 900
sub_results = results[results["area"] == area_plot]
x = sub_results['gps_size'].to_numpy()
y = sub_results['total_py_par'].to_numpy()

plt.plot(x, y, 'go-', linewidth = 2, color = 'blue', label = "Python (Parallel)", marker = '*', markersize = 10)

y = sub_results['total_py'].to_numpy()

plt.plot(x, y, 'go-', linewidth = 2, color = 'magenta', label = "Python", marker = '^', markersize = 10)

y = sub_results['total_pg_qgis'].to_numpy()

plt.plot(x, y, 'go-', linewidth = 2, color = 'green', label = "Postgis+QGIS", marker = 'o', markersize = 10)

y = sub_results['total_qgis'].to_numpy()

plt.plot(x, y, 'go--', linewidth = 2, color = 'red', label = "QGIS", marker = 'v', markersize = 10)

plt.legend()
plt.grid(linestyle = "--", linewidth = 0.5)
plt.title("Test Area " + str(area_plot) + " $km^2$: Total Processing Time", fontweight="bold")
plt.xlabel("GPS Point Count (millions)")
plt.ylabel("Total Time (s)")
plt.savefig("area_900_vs_gps_all.png", dpi = 600)
```
